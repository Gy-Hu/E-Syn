```python
import xgboost as xgb
import pandas as pd
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt

#X, y = load_breast_cancer(return_X_y=True)

# import csv file: 
df = pd.read_csv('/data/guangyuh/coding_env/E-Brush/xgboost_reg/fuzz_circuit_analysis_small_size.csv')
df['area_delay_product'] = 0.6*df['delay'] + 0.4*df['area']
# X is the first 30 columns, y is the last column
X = df.iloc[:, :8].values
y = df.iloc[:, -1].values

#df = pd.DataFrame(X, columns=range(30))
#df['y'] = y

model = xgb.XGBRegressor()
model.fit(X, y)

importances = model.feature_importances_
#importances = pd.Series(importances, index=range(X.shape[1]))
importances = pd.Series(importances, index=df.columns[:8])
importances.plot.bar()
```
