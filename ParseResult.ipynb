```python
import os
import re
import pandas as pd

# Define the directory
directory = "/data/guangyuh/coding_env/E-Brush/result_bak"

# Initialize an empty dictionary to store the results
results = {}

# Iterate over the files in the directory
for filename in os.listdir(directory):
    # Check if the file is .txt file
    if filename.endswith(".txt"):
        # Define the full file path
        filepath = os.path.join(directory, filename)

        # Open & read the file
        with open(filepath, 'r') as file:
            lines = file.readlines()

            # Initialize an empty dictionary to store the file's results
            file_results = {'original_circuit': {}, 'optimized_circuit': {}}

            # Iterate over the lines in the file
            for line in lines:
                flag_original = False; flag_optimized = False
                # Check if the line contains the circuit information, and happens in the first time
                if 'original_circuit' in line and not flag_original:
                    circuit = 'original_circuit' ; flag_original = True
                elif 'optimized_circuit' in line and not flag_optimized:
                    circuit = 'optimized_circuit'; flag_optimized = True
                
                

                if match_power := re.search(r'power = (\d+\.\d+)', line) or re.search(r'power =(\d+\.\d+)', line):
                    power = float(match_power[1])
                    file_results[circuit]['power'] = power

                match_gates = re.search(r'Gates =\s*([\d]+)', line)
                match_cap = re.search(r'Cap =\s*([\d\.]+)', line)
                match_area = re.search(r'Area =\s*([\d\.]+)', line)
                match_delay = re.search(r'Delay =\s*([\d\.]+)', line)

                # if match_gates and match_cap and match_area and match_delay all not None
                if (match_gates is not None and match_cap is not None and match_area is not None and match_delay is not None):
                    gates = int(match_gates[1])
                    cap = float(match_cap[1])
                    area = float(match_area[1])
                    delay = float(match_delay[1])
                    file_results[circuit].update({'gates': gates, 'cap': cap, 'area': area, 'delay': delay})

            

        # Store the file's results
        results[filename.rsplit('.', 1)[0]] = file_results

'''
print(results)
# Create dataframe for original_circuit
df_original = pd.DataFrame(results['adder']['original_circuit'], index=[0])
# Create dataframe for optimized_circuit
df_optimized = pd.DataFrame(results['adder']['optimized_circuit'], index=[0])
print(f"----------------------------------Original Circuit: ----------------------------------")
print(df_original)
print(f"----------------------------------Optimized Circuit----------------------------------")
print(df_optimized)
'''

# for every circuit in results, parse the results into a dataframe and print

for circuit in results:
    df_original = pd.DataFrame(results[circuit]['original_circuit'], index=[0])
    df_optimized = pd.DataFrame(results[circuit]['optimized_circuit'], index=[0])
    print(f"----------------------------------{circuit}----------------------------------")
    print(f"----------------------------------Original Circuit: ----------------------------------")
    # print with no index
    #print(df_original.to_string(index=False))
    print(df_original)
    print(f"----------------------------------Optimized Circuit----------------------------------")
    #print(df_optimized.to_string(index=False))
    print(df_optimized)
```
